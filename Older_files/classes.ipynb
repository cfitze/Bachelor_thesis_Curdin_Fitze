```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

#Q: change pandas plotting backend to plotly
pd.options.plotting.backend = "plotly"
```

```python
class Messreihe():
    def __init__(self, name, filename):
        self.name = name
        self.data = []
        self.filename = filename

    def read_excel(self):
        self.data = pd.read_excel(self.filename, sheet_name="15min")
```

```python
filename = "BA_23FS_Curdin_Fitze_5_7_9_11_13_TSextract.xlsx"
messreihe = Messreihe("Testfile", filename)
```
